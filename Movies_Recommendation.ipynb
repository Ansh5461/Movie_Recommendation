```python
#This extracts the name of the director
def get_director(x):
    for i in x:
        if i["job"] == "Director":
            return i["name"]
    return np.nan

```

```python
def get_list(x):
    if isinstance(x, list): #if x is a list
        names = [i["name"] for i in x]
        
        #what names will do is store something, what is that something
        #its i["name"] where i is iterating over x
        
        if len(names) > 3:
            names = names[:3]
            
        return names
    
    return []
```

```python
df2["director"] = df2["crew"].apply(get_director)

features = ["cast", "keywords", "genres"]

for feature in features:
    df2[feature] = df2[feature].apply(get_list)
    
    #we will pass each column to get_list, and if column is a list, which it is, names will iterate over it
    #and return just top 3 elements
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ",""))  for i in x]
    
    else:
        if isinstance(x, str):
            return [str.lower(x.replace(" ",""))]
        else:
            return ""
        
features = ["director", "cast", "keywords", "genres"]
for feature in features:
    df2[feature] = df2[feature].apply(clean_data)
```

```python
# Now, let’s create a “soup” containing all of the metadata information extracted to input into the vectorizer.
# We want to predict movies by joining their main features to 1 feature and passing it to vectorizer

def create_soup(features):
    #print(type(features['keywords']))
    return ' '.join(features['keywords']) + ' ' + ' '.join(features['cast']) + ' ' + ' '.join(features['director'])+ ' ' + ' '.join(features['genres'])


df2["soup"] = df2.apply(create_soup, axis=1)
print(df2["soup"].head())
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity


model = CountVectorizer(stop_words="english")
matrix = model.fit_transform(df2["soup"])
print(matrix.shape)

cosine_sim2 = cosine_similarity(matrix, matrix) 
print(cosine_sim2.shape)

df2 = df2.reset_index()
indices = pd.Series(df2.index, index=df2['title_x'])
```

```python
"""
Our movie recommendation engine works by suggesting movies to the user based on the metadata information. 
The similarity between the movies is calculated and then used to make recommendations. For that, our text data 
should be preprocessed and converted into a vectorizer using the CountVectorizer. As the name suggests, 
CountVectorizer counts the frequency of each word and outputs a 2D vector containing frequencies.

We don’t take into account the words like a, an, the (these are called “stopwords”) because these words are 
usually present in higher amounts in the text and don’t make any sense.

There exist several similarity score functions such as cosine similarity, Pearson correlation coefficient, etc. 
Here, we use the cosine similarity score as this is just the dot product of the vector output by the CountVectorizer.

We also reset the indices of our dataframe.



Create a reverse mapping of movie titles to indices. By this, we can easily find the title of the movie based on the index.

"""


indices = pd.Series(df2.index, index=df2["title_x"]).drop_duplicates()

print(indices.head())
```

```python
"""

Get the index of the movie using the title.
Get the list of similarity scores of the movies concerning all the movies.
Enumerate them (create tuples) with the first element being the index and the second element is the cosine similarity score.
Sort the list of tuples in descending order based on the similarity score.
Get the list of the indices of the top 10 movies from the above sorted list. Exclude the first element because it is the 
title itself.
Map those indices to their respective titles and return the movies list.

"""


def get_recommendations(title, cosine_sim=cosine_sim2):
    idx = indices[title]
    similarity_scores = list(enumerate(cosine_sim[idx]))
    similarity_scores= sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    similarity_scores= similarity_scores[1:11]
    # (a, b) where a is id of movie, b is similarity_scores
    movies_indices = [ind[0] for ind in similarity_scores]
    movies = df2["title_x"].iloc[movies_indices]
    return movies


print("################ Content Based System #############")
print("Recommendations for The Dark Knight Rises")

print(get_recommendations("The Dark Knight Rises", cosine_sim2))
print()
print("Recommendations for Avengers")
print(get_recommendations("The Avengers", cosine_sim2))
```
